# Funtion for normalizing texts

 http://agailloty.rbind.io/project/nlp_clean-text/*italicized text*

```python

def clean_text(text):
    """
    This function takes as input a text on which several
    NLTK algorithms will be applied in order to preprocess it
    """
    tokens = word_tokenize(text)
    # Remove the punctuations
    tokens = [word for word in tokens if word.isalpha()]
    # Lower the tokens
    tokens = [word.lower() for word in tokens]
    # Remove stopword
    tokens = [word for word in tokens if not word in stopwords.words("english")]
    # Lemmatize
    lemma = WordNetLemmatizer()
    tokens = [lemma.lemmatize(word, pos = "v") for word in tokens]
    tokens = [lemma.lemmatize(word, pos = "n") for word in tokens]
    return tokens

def clean_text_including_stopwords(text):
    """
    This function takes as input a text on which several
    NLTK algorithms will be applied in order to preprocess it
    """
    tokens = word_tokenize(text)
    # Remove the punctuations
    tokens = [word for word in tokens if word.isalpha()]
    # Lower the tokens
    tokens = [word.lower() for word in tokens]

    # Lemmatize
    lemma = WordNetLemmatizer()
    tokens = [lemma.lemmatize(word, pos = "v") for word in tokens]
    tokens = [lemma.lemmatize(word, pos = "n") for word in tokens]
    return tokens
```

# First Method: Lexical Window

In this method, we need the positions of words in the documents and should avoid removing stopwords.

## Calculating and Updating Lexical Windows for All Document Words

We consider a radius of 2, resulting in a window length of 5.

If a word appears at the beginning of a sentence or has fewer than two words before or after it, fewer words need to be checked. This case has also been addressed in this section of the code.

Approximately 10 seconds for 500 documents.

```python
wr = 2 # window radius
windows = {}
def default_vector():
  return [0.00000001]*len(voclist)

for doc in docs_matris_sent:
  for sent in doc:
    for index, word in enumerate(sent):
      if word not in windows.keys():
        windows[word] = default_vector()
      lr, rr = min(index, wr), min(len(sent) - index -1, wr)
      for i in range(1, lr):
        windows[word][voclist.index(sent[index - i])] += 1
      for i in range(1, rr):
        windows[word][voclist.index(sent[index + i])] += 1
print(windows['post'])
```

## Computing Similarity Matrix for All Documents
This matrix equals the sum of the lexical window matrices of all words divided by their count.

For this calculation, the mean function from the NumPy library has been utilized.
Subsequently, we have normalized this matrix.

Approximately 20 seconds for 500 documents.

```python
similarityDocList =[]
for doc in docs_matris:
  # normalizing vector in: https://stackoverflow.com/questions/21030391/how-to-normalize-a-numpy-array-to-a-unit-vector
  x = np.mean([windows[w] for w in doc], axis=0)

  x = x / np.linalg.norm(x)
  similarityDocList.append(x)

print( max(similarityDocList[3].tolist()))
```

# Query Search Function for the First Method

## Part One: Computing Similarity Matrix for the Query

## Part Two: Cosine Comparison of the Query Similarity Matrix and the Document Matrix

## Part Three: Sorting the Attained Information Based on Similarity Scores

```python

def search_query_R1(query):
# create similarity matris for query 1
  clean_query = clean_text(query)
  query_similarity = np.mean([windows[w] if w in windows.keys() else default_vector() for w in clean_query], axis=0)
  query_similarity = (query_similarity / np.linalg.norm(query_similarity))
# similarity matris has been created 1

# cosin similarity 2
  sims = []
  for index, vec in enumerate(similarityDocList):
    cos_sim = np.dot(query_similarity, vec)/(np.linalg.norm(query_similarity)*np.linalg.norm(vec))

    if not np.isnan(np.sum(cos_sim)):
      sims.append((index, float(cos_sim)))

# end cosin similarity 2
# sort 3
  sims.sort(key=lambda x: x[1], reverse=True)

  return sims
# endsort 3


```

## Second Method - Sentence-based Lexical Similarity
Similar to the first method, with the difference that the length of the lexical window is equal to the length of the sentence.

12 min

```python
sent_windows = {}
def default_vector():
  return [0.00000001]*len(voclist)

for doc in docs_matris_sent:
  for sent in doc:
    for index, word in enumerate(sent):
      if word not in sent_windows.keys():
        sent_windows[word] = default_vector()
      for w in sent:
        if w != word:
          if w in voclist:
            sent_windows[word][voclist.index(w)] += 1
print(sent_windows['post'])
```

## Third Method: Character-based Approach

In this method, we measure the similarity of words with similar characters in each document to every query word. This measurement will be done using the **Jaccard** method.

Subsequently, after constructing a matrix for all words present in the document and calculating their Jaccard similarity, we perform the following steps for both documents and queries:

1. Sum the matrices for all words based on their constituents.
2. Calculate the average of these matrices.

Then, we calculate cosine similarity and sort the results.

```python
def jaccard(intersection,union):
  return float(intersection)/union
```

```python
def determine_score_letter_two_word_using_jaccard(queryWord,docWord):       #determine score between two word first one being query word
    setOfLetterQueryWord = set()
    setOfLetterDocWord = set()

    for i in queryWord:
      setOfLetterQueryWord.add(i)              #letter in query`s words but just once
    for i in docWord:
      setOfLetterDocWord.add(i)                #letter in doc`s words but just once

    RepetitiousOFQueryWord = len(queryWord) - len(setOfLetterQueryWord)         #number of repeated letter in query`s word
    RepetitiousOFDocWord = len(docWord) - len(setOfLetterDocWord)               #number of repeated letter in doc`s word

    sameLetterCounter=0
    for i in setOfLetterDocWord:
      if i in setOfLetterQueryWord:
        sameLetterCounter += 1

    intersection= sameLetterCounter
    union=  len(setOfLetterQueryWord) + len(setOfLetterDocWord)

    jaccardScore= jaccard(intersection,union)

    return jaccardScore
```

## Fourth Method: Subjective Comparison Function Based on Query using Jaccard Method

### Identical to the Third Method, with the distinction that we employ a different function to calculate the similarity between two words.

```python
def determine_score_rhymes_two_word_using_jaccard(queryWord,docWord):       #determine score between two word first one being query word

    lastIndex=0
    for i in range(len(docWord)+1):

      if i==0:
        i = 1     # we need inverse counter
      if i > len(queryWord):
        break
      if docWord[-i]==queryWord[-i]:
        pass
      else:
        lastIndex=i-1         #last one doest count, it`s not same as query
        break

    intersection= lastIndex
    union=  len(docWord) + len(queryWord) - 2*intersection

    jaccardScore= jaccard(intersection,union)

    return jaccardScore
```
